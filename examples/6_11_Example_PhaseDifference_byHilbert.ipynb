```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert

# Generate two signals with a phase difference
t = np.linspace(0, 1, 1000)
signal1 = np.sin(2 * np.pi * 10 * t)
signal2 = np.sin(2 * np.pi * 10 * t + np.pi/4)  # Phase difference of pi/4

# Compute the Hilbert transform of each signal
analytic_signal1 = hilbert(signal1)
analytic_signal2 = hilbert(signal2)

# Extract the instantaneous phase of each signal
phase1 = np.unwrap(np.angle(analytic_signal1))
phase2 = np.unwrap(np.angle(analytic_signal2))

# Calculate the phase difference
phase_diff = phase2 - phase1

# Plot the signals and the phase difference
plt.figure()
plt.plot(t, signal1, label='Signal 1')
plt.plot(t, signal2, label='Signal 2')
plt.legend()

plt.figure()
plt.plot(t, phase_diff)
plt.xlabel('Time')
plt.ylabel('Phase Difference')
plt.show()
```

Explanation:
1. Hilbert transform:
The Hilbert transform creates an analytic signal, which allows you to extract the instantaneous phase and amplitude of a signal.
2. Unwrap:
The np.unwrap function unwraps the phase to avoid jumps caused by the circular nature of phase angles.
3. Phase difference:
The difference between the unwrapped phases represents the phase difference between the two signals.
