```python
import numpy as np

# Create a complex-valued matrix
A = np.array([[1+2j, 2-1j], [3+1j, 4-2j]])

# Perform SVD
U, s, Vh = np.linalg.svd(A)

# Reconstruct the original matrix
S = np.diag(s)
A_reconstructed = U @ S @ Vh

print("Original matrix:")
print(A)
print("Singular values:")
print(s)
print("Reconstructed matrix:")
print(A_reconstructed)
```

```python
# Example set a complex
x = np.array([1,2,3,4,5])

# np.complex(x, x)
complex_num = np.array([complex(z, z) for z in x])
print(complex_num)
```

```python
def energy(signal):
    return np.sum(np.abs(signal)**2)
energy(complex_num)
```
