```python
# Probalility of no transmitter occupies a slot depends on the number of transmitters~$N$ and the number of slots~$D$.
# The top left is~$P_0$ (no transmission in a slot),
# right is~$P_1$ (one transmission),
# the bottom left is~$P_2$ (two transmitters collide),
# and right is~$P_3$ (three and more transmitters collide).}

import numpy as np
import matplotlib.pyplot as plt
```

```python
Qs = np.array([4,5,6,7,8,9,10])
#Ns = range(50,1000,100)
Ns = np.linspace(50., 1000., 55)

#print(Qs)
#print(Ns)
for Q in Qs:
  Pr = np.empty(len(Ns))
  for N, i in zip(Ns, range(len(Ns))):
    P0 = (((2**Q-1)/2**Q)**N) # P0
    P1 = (N * (1/(2**Q))) *  (((2**Q-1)/2**Q)**(N-1)) # P1
    P2 = ((N*(N-1))/2) * (1/(2**(2*Q))) * (((2**Q-1)/2**Q)**(N-2)) # P2
    P3 = 1 - P0 - P1 - P2
    P = P3
    Pr[i] = P
  plt.plot(Ns, Pr, label = f'Q = {Q}')

plt.rcParams['font.family'] = 'DejaVu Serif'
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['xtick.labelsize'] = 12#24
plt.rcParams['ytick.labelsize'] = 12#24
plt.rcParams['legend.fontsize'] = 12#24
plt.rcParams['axes.labelsize'] = 16
plt.xlabel('Number of transmitters, N', fontname='DejaVu Serif')
plt.ylabel('Probability', fontname='DejaVu Serif')
plt.legend()
plt.show()
```
